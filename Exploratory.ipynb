### Grouping Data using groupby() method

```python
df_test = df[['drive-wheels','body-style','price']]
df_grp = df_test.groupby(['drive-wheels','body-style'], as_index = False).mean()
df_grp
```

### As above table is not that compatible from the point of drawing insights by grouping the objects, pivot table is a better option to bring out visuals for grouped data.

```python
df_pivot = df_grp.pivot(index="drive-wheels",columns ="body-style")
df_pivot
```

## Another way to visualize Pivot Tables is using Heat Maps, Heat Maps are used to visualize or plot Targeted Variable with Multiple Variables.

```python
sns.heatmap(df_pivot,cbar_kws={'label': 'price'},cmap="RdBu")
```
