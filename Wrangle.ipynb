# Data Wrangling of Automobile DataSet

```python
import pandas as pd 
import numpy as np
import seaborn as sns
url_csv = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv"
df = pd.read_csv(url_csv)
df
```

# Replacing Missing Data

Replace '?' with NaN using .replace() method which is a builtin function in Pandas used to replace Values along the dataframe.

```python
import numpy as np

df.replace(np.nan,0, inplace = True)
df
```

# Evaluating Missing Data 

There are two methods of doing so:
1. isnull()
2. notnull()

and it returns a boolean value.

```python
missing_data = df.isnull()
missing_data
```

```python
Not_Null_data = df.notnull()
Not_Null_data
```

Basically this dataset has no null values and so the isnull() and notnull() funtions return False and True respectively for all the rows and columns in dataset.

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

For a missing Value the value_counts() method will return True and False for no missing value.

# Replacing Missing data with Mean

```python
mean_stroke = df["stroke"].astype("float").mean(axis=0)
df["stroke"].replace(" ",mean_stroke, inplace = True)
print(mean_stroke)
df
```

Replace NaN / Blank Values in a column with the Mean of that Column values.

```python
df["horsepower-binned"].replace(0,"High",inplace = True)
df
```

# Data Standardization/ Formatting

Different Agencies use different formats or measures for a particular value(column), so data formatting plays an important role there.

```python
df["city-L/100km"] = 235 / df["city-mpg"]
df.head()
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df
```

# Data Normalization

Transforming numeric columns to a standard scale. Generally used for higher values that differs from others multiple times.
Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1

```python
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df['height'] = df['height']/df['height'].max() 
df[["length","width","height"]].head()
```

Normalized length,width and heigth are in the range [0,1]
