# DATA VISUALIZATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('eda.csv')
```

```python
df.head()
```

```python
cols = ['Unnamed: 0', 'Unnamed: 0.1']
df = df.drop(cols, axis = 1,inplace = True)
```

```python
df
```

```python
df.columns
```

### Creating Visuals and see how Price is being affected by other features/components of the dataset.

```python
width = 12
height = 10
plt.figure(figsize = (width,height))


plt.scatter('highway-L/100km','price',color = 'g', data = df)
plt.xlabel('highway-L/100km')
plt.ylabel('price')
plt.legend()

plt.title('highway-L/100km vs price')
plt.show()
```

```python
width = 12
height = 10
plt.figure(figsize = (width,height))

plt.barh('city-mpg', 'price', color = 'r', data =df)
plt.xlabel('price')
plt.ylabel('city-mpg')

plt.show()
```

```python
width = 12
height = 10
plt.figure(figsize = (width,height))
plt.scatter('peak-rpm','price',color = 'blue', data = df)

plt.title('peak-rpm vs price')
plt.show()
```

```python
width = 12
height = 10
plt.figure(figsize = (width,height))

plt.scatter('compression-ratio','price', color = 'red', data = df)
plt.xlabel('compression-ratio')
plt.ylabel('price')

plt.show()
```

```python
width = 12
height = 10
plt.figure(figsize = (width,height))

plt.scatter('stroke', 'price', color = 'black', data = df)
plt.xlabel('stroke')
plt.ylabel('price')
plt.show()
```
